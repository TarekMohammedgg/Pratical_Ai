## Data Cleaning 


```python
import pandas as pd 
df = pd.read_excel('dataset/Customer Call List.xlsx') 
for column in df.columns : 
    print(f'{column} : {df[column].unique()} ')
    print("\n")
    
```

```python
df.isnull().sum()

df.iloc[8, : ]
```

```python
#drop duplicated values 
df = df.drop_duplicates() 
df 
```

```python
#drop unusefull columns 
df.drop(columns = 'Not_Useful_Column' , inplace= True ) 
#inplace used for automaticly save the new settings on the original datafram without use df = new df version 
```

```python
#strip قص / تجريد  الاجزاء الغير مفهومة بداخل الاسماء 

df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
df
```

```python
#solve problem in phone_Number 
df['Phone_Number'] =  df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '' , regex= True ) #must use regex to ignor warrning or errors because non-values 'nan'  
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x)) #convert each number to string to can تقطيع number to slices 
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3]+ '-' + x[3:6] + '-' + x[6:9]) ## splitting numbers with '-' 
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '').replace('Na--' , '') 

df['Phone_Number']
```

```python
#splitting the address to be easy to read the values 
df[['Street Adress', 'State'  , 'Zip_code']] = df['Address'].str.split(',' ,  expand= True )
df
```

```python
# in paying customer and not contacting we want to standard the format of two columns 
df['Paying Customer'] = df['Paying Customer'].replace('Yes' , 'Y').replace("No" , "N") 
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes' , 'Y').replace('No' , 'N' )
df
```

```python
# fill null values 

df = df.fillna('') 
```

```python
df
```

```python
# if the values of column 'Do_Not_Contact' is  Y drop this row 
for x in df.index : 
    if df.loc[x  ,  'Do_Not_Contact'] == 'Y' : 
        df.drop(x , inplace= True )  
        
        
df 
```

```python
# do the same somthing on 'phone_Number' column but when the row is blanc 

for x in df.index : 
    if df.loc[x , 'Phone_Number'] == "" : 
        df.drop(x , inplace= True )

## another way to drop null values 
df.dropna(subset= 'Phone_Number'  , inplace= True ) 
df 
```

```python
# reset the index 

df = df.reset_index(drop = True )
```

```python
df = df.drop(columns= ['Do_Not_Contact'  , 'Address'  ]) 
df 
```
